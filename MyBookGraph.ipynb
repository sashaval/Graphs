# Create the Plot

## Non-cumulative Plot

```r
options(repr.plot.width = 10, repr.plot.height = 5)

ggplot(my_data, aes(x = mon, y = MahBook)) +
  geom_line(colour = "black", alpha = 1) + geom_point(size = 2) +
  facet_grid(scales = "free_x", space = "free_x") +
  geom_smooth(
    method = "lm",
    se = FALSE,
    colour = "forest green"
    ) +
  scale_y_continuous(breaks = seq(0, 205000, by = 25000)) +
  scale_x_continuous(breaks = seq(2, 12, by = 2)) +
  labs(
    x = "Month",
    y = "Word Count",
    title = "Manuscript Word Count"
    ) +
  theme(
    axis.title = element_text(size = 13),
    plot.title = element_text(size = 15),
    axis.text = element_text(size = 11),
   )
```
